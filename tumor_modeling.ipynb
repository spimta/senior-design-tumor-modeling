```python
# 3D tumor simulation

import numpy as np
import math
import matplotlib.pyplot as plt
import random
from mpl_toolkits import mplot3d
        

# print spatial distribution
def print_spatial(x, y, CLT, t, arg='none'):
    if arg == 'live':
        for x,y,CLT in zip(x,y,CLT):
            if CLT <= 0:
                continue

            plt.scatter(x,y,c="red")

        plt.title("spatial distributions of live cell at time "+str(t))
        plt.show()
    
    else:
        for x,y,CLT in zip(x,y,CLT):
            color = "red"
            if CLT <= 0:
                color = "black"

            plt.scatter(x,y,c=color)

        plt.title("spatial distributions of all cells at time "+str(t))
        plt.show()
    
    return


# print population time dependencies
def print_population(pop_live, pop_death, times):
    plt.title("population")
    plt.plot(times, pop_live, c="red")
    plt.plot(times, pop_death, c="black")
    plt.xlabel("time")
    plt.ylabel("number of cells")
    plt.show()


# initialization
x_acc = [0]
y_acc = [0]
z_acc = [0]
x_spd = [0]
y_spd = [0]
z_spd = [0]
x = [0]
y = [0]
z = [0]

TOS = 150 # time of simulation
mass = 10 # cell mass
r_cell = 1 # cell radius
p_death = 0.005 # probability of death
p_div = 0.0625 # probablity of division
lifetime = 30 # lifetime for each cell
alpha = 0.005 # coefficient alpha for haptotaxis
beta = 0.01 # coefficient beta for adhension

sigma = 2

CLT = [lifetime] #initial 1 cell

population_live = []
population_dead = []
times = []

for t in range(0,TOS):
    
    # Cell's division process
    for i in range(0, len(CLT)):
        
        # if current cell is alive
        if CLT[i]>0:
            tmp = random.random() # roll a die

            # cell death
            if tmp<p_death:
                CLT[i] = 0

            # cell division
            elif tmp>=1-p_div:
                CLT.append(random.gauss(lifetime, sigma))
                psi = np.random.rand() #random value for angle
                phi = np.random.rand()
                x.append(2*r_cell*math.cos(2*math.pi*psi)*math.sin(2*math.pi*phi)) # initial x axis
                y.append(2*r_cell*math.sin(2*math.pi*psi)*math.sin(2*math.pi*phi)) # initial y axis
                z.append(2*r_cell*math.cos(2*math.pi*phi))
                x_acc.append(0.0)
                y_acc.append(0.0)
                z_acc.append(0.0)
                x_spd.append(0.0)
                y_spd.append(0.0)
                z_spd.append(0.0)
        
    
    # cell's motion
    for j in range(0,len(CLT)):
        
        # only living cell has cell's motion
        if CLT[j]>0:
            
            # haptotaxis force
            c_out = 0
            c_in = 0
            
            rj = x[j]**2+y[j]**2+z[j]**2
            # count number of innner and outer cells
            for k in range(0,len(CLT)):
                if CLT[k]>0:
                    rk = x[k]**2+y[k]**2+z[k]**2
                    if rk > rj:
                        c_out += 1
                    elif rk < rj:
                        c_in += 1
                    
            rj = math.sqrt(rj)
            Fx = -alpha*(c_out-c_in)*x[j]/(rj+0.01)
            Fy = -alpha*(c_out-c_in)*y[j]/(rj+0.01)
            Fz = -alpha*(c_out-c_in)*z[j]/(rj+0.01)
            
            # adhension force for j-th cell
            for i in range(0,len(CLT)):
                if i != j and CLT[i]>0:
                    dx = x[j]-x[i]
                    dy = y[j]-y[i]
                    dz = z[j]-z[i]
                    dr = math.sqrt(dx**2+dy**2+dz**2)
                    Fx += -beta*dx/(dr+1)**3*math.exp(-dr/(4*r_cell))
                    Fy += -beta*dy/(dr+1)**3*math.exp(-dr/(4*r_cell))
                    Fz += -beta*dz/(dr+1)**3*math.exp(-dr/(4*r_cell))
                    
            x_acc[j] = Fx / mass
            y_acc[j] = Fy / mass
            z_acc[j] = Fz / mass
            
            x_spd[j] += x_acc[j]
            y_spd[j] += y_acc[j]
            z_spd[j] += z_acc[j]
            
            x[j] += x_spd[j]
            y[j] += y_spd[j]
            z[j] += z_spd[j]
            
        CLT[j] -= 1
        
    for n in range(0,301):
        if n == 300*t/TOS:
            live_counter = 0
            dead_counter = 0
            for k in range(0,len(CLT)):
                if CLT[k]>0:
                    live_counter += 1
                else:
                    dead_counter += 1
            
            population_live.append(live_counter)
            population_dead.append(dead_counter)
            
            times.append(300*t/TOS)
        
fig = plt.figure()
ax = plt.axes(projection='3d')

for x,y,z,CLT in zip(x,y,z,CLT):
    color = "red"
    if CLT <= 0:
        color = "black"
        
    ax.scatter3D(x,y,z,c=color)
    
```
